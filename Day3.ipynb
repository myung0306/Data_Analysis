```python
print(df.describe())
print("")
"""
1. 평균
2. 각 값에서 평균 빼기
3. 제곱 후 전체 개수 만큼 나누기
4. 분산에 루트를 씌우면 표준편차
"""
print(df.describe(include='all'))
#unique:고유값
#top:최고 많이 나온 수(고유값이 없으므로 최빈값 구하기 어려움)
#freq:최빈값 빈도
```

```python
#데이터의 개수
print(df.count())
```

```python
#데이터의 개수
unique_values=df.origin.value_counts()
print(unique_values)
print("")
print(type(unique_values))
```

```python
#평균값(문자열 값은 빠지고 수치형 데이터만)
#Nan값이나 문자열 데이터가 하나만 있어도 통계 계산이 불가(문자열의 영향을 받음)
#숫자가 문자로 맞춰줌
print(df.mean())
print("")
print(df.mpg.mean()) #단일적인 평균값
print("")
print(df[['mpg', 'weight']].mean())
```

```python
#중간값
print(df.median())
print("")
print(df.mpg.median())
print("")
print(df[['mpg', 'weight']].median())
```

```python
#최대값
print(df.max())
print("")
print(df.mpg.max())
print("")
print(df[['mpg', 'weight']].max())
```

```python
#최소값
print(df.min())
print("")
print(df.mpg.min())
print("")
print(df[['mpg', 'weight']].min())
```

```python
#표준편차
print(df.std())
print("")
print(df.mpg.std())
print("")
print(df[['mpg', 'weight']].std())
```

```python
#상관계수:얼마나 연관성이 있는지 여부
#선이 눈에 올라가고 내려가는게 뚜렷하게 보이면 상관관계가 강하다
#선이 퍼지면 상관관계가 약하다
#0에 가까울 수록 데이터는 퍼짐, 1과 -1에 가까울 수록 데이터 오밀조밀
print(df.corr())
print("")
print(df[['mpg', 'weight']].corr())
```

```python
#데이터프레임 생성
import pandas as pd

df=pd.read_excel("남북한발전전력량.xlsx")
#ffill:NaN을 위의 데이터로 대체
df.fillna(method='ffill', inplace=True)
print(df)
```

```python
df_ns=df.iloc[[0, 5], 2:]
print(df_ns)
```

```python
df_ns.index=['South', 'North']
print(df_ns)
```

```python
df_ns.columns=df_ns.columns.map(int)
print(df_ns)
print("")
df_ns.plot()
```

```python
tdf_ns=df_ns.T
print(tdf_ns)
tdf_ns.plot()
```

```python
import pandas as pd

df = pd.read_excel("남북한발전전력량.xlsx")
df.fillna(method='ffill', inplace=True)

df_ns=df.iloc[[0, 5], 2:]
df_ns.index=['South', 'North']

df_ns.columns=df_ns.columns.map(int)
tdf_ns=df_ns.transpose()
tdf_ns.plot(kind='bar')
```
