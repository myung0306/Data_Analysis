```python
#데이터의 개수
unique_values=df.origin.value_counts()
print(unique_values)
print("")
print(type(unique_values))
```

```python
#평균값(문자열 값은 빠지고 수치형 데이터만)
#Nan값이나 문자열 데이터가 하나만 있어도 통계 계산이 불가(문자열의 영향을 받음)
#숫자가 문자로 맞춰줌
print(df.mean())
print("")
print(df.mpg.mean()) #단일적인 평균값
print("")
print(df[['mpg', 'weight']].mean())
```

```python
#중간값
print(df.median())
print("")
print(df.mpg.median())
print("")
print(df[['mpg', 'weight']].median())
```

```python
#최대값
print(df.max())
print("")
print(df.mpg.max())
print("")
print(df[['mpg', 'weight']].max())
```

```python
#최소값
print(df.min())
print("")
print(df.mpg.min())
print("")
print(df[['mpg', 'weight']].min())
```

```python
#표준편차
print(df.std())
print("")
print(df.mpg.std())
print("")
print(df[['mpg', 'weight']].std())
```

```python
#상관계수:얼마나 연관성이 있는지 여부
#선이 눈에 올라가고 내려가는게 뚜렷하게 보이면 상관관계가 강하다
#선이 퍼지면 상관관계가 약하다
#0에 가까울 수록 데이터는 퍼짐, 1과 -1에 가까울 수록 데이터 오밀조밀
print(df.corr())
print("")
print(df[['mpg', 'weight']].corr())
```

```python
"""
kind 인자
line : 선 그래프(추세)
bar : 수직 막대 그래프
barh : 수평 막대 그래프
his : 히스토그램(기본으로 수직 막대 그래프)
box : 박스플롯(4분위수를 그래프로 도식화)
kde : 커널 밀도 그래프(히스토그램을 선처럼 표시)
area : 면적 그래프(추세의 합계)
pie : 파이 그래프
scatter : 산점도 그래프
"""
```

```python
#막대그래프는 비교
import pandas as pd

df = pd.read_excel("남북한발전전력량.xlsx")
df.fillna(method='ffill', inplace=True)

df_ns=df.iloc[[0, 5], 2:]
df_ns.index=['South', 'North']

df_ns.columns=df_ns.columns.map(int)
tdf_ns=df_ns.transpose()
tdf_ns.plot(kind='bar')
```

```python
#히스토그램은 빈도
import pandas as pd

df=pd.read_excel("남북한발전전력량.xlsx")
df.fillna(method='ffill', inplace=True)

df_ns=df.iloc[[0, 5], 2:]
df_ns.index=['South', 'North']

df_ns.columns=df_ns.columns.map(int)
tdf_ns=df_ns.transpose()
#pd.to_numeric-데이터 값을 문자열을 숫자여로 변경
tdf_ns=tdf_ns.apply(pd.to_numeric)
tdf_ns.plot(kind='hist')
```

```python
#axe 객체
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family']='NanumBarunGothic'

df=pd.read_excel('시도별 전출입 인구수.xlsx')
df=df.fillna(method='ffill')
a=(df['전출지별']=='서울특별시')&(df['전입지별']!='서울특별시')
df_seoul=df[a]
df_seoul=df_seoul.drop(['전출지별'], axis=1)
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.set_index('전입지', inplace=True)
sr_gyounggi=df_seoul.loc['경기도']

plt.style.use('ggplot')

fig=plt.figure(figsize=(10, 10))
ax1=fig.add_subplot(2, 1, 1)
ax2=fig.add_subplot(2, 1, 2)

ax1.plot(sr_gyounggi, 'o', markersize=10)
ax2.plot(sr_gyounggi, marker='o', markersize=10, markerfacecolor='green',
         color='olive', linewidth=2, label='서울->경기')
ax2.legend(loc='best')

ax1.set_ylim(50000, 700000)
ax2.set_ylim(50000, 700000)

ax1.set_xticklabels(sr_gyounggi.index, rotation=75)
ax2.set_xticklabels(sr_gyounggi.index, rotation=75)

plt.show()
```
