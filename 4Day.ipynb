```python
#버블 차트
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel("auto-mpg.xlsx", header=None)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'moder_year', 'origin', 'name']

plt.style.use("default")

cylinders_size=df.cylinders / df.cylinders.max() * 300

df.plot(kind='scatter', x='weight', y='mpg', color='coral', s=cylinders_size, figsize=(10, 5),
        alpha=0.3) #alpha-점의 투명도
plt.title("Scatter Plot - mpg vs weight")
plt.show()
```

```python
#파이 차트
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("auto-mpg.xlsx", header = None)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'moder_year', 'origin', 'name']

plt.style.use("default")

df['count']=1
df_origin=df.groupby('origin').sum()

df_origin.index=['USA', 'EU', 'JAPAN']
df_origin['count'].plot(kind='pie', figsize=(7, 5), autopct="%0.1f%%",
                        startangle=10, colors=['chocolate', 'bisque', 'cadetblue'])

plt.title("Model Origin", size=20)
plt.axis('equal') #파이를 원으로 만듦
plt.legend(loc='best') #범례

plt.show()
```

```python
#박스 플롯
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'NanumBarunGothic'

df = pd.read_excel("auto-mpg.xlsx", header = None)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'moder_year', 'origin', 'name']

plt.style.use("seaborn-poster")

fig=plt.figure(figsize=(15, 5))
ax1=fig.add_subplot(1, 2, 1)
ax2=fig.add_subplot(1, 2, 2)

ax1.boxplot(x=[df[df['origin']==1]['mpg'],
               df[df['origin']==2]['mpg'],
               df[df['origin']==3]['mpg']],
            labels=['USA', 'EU', 'JAPAN'])

ax2.boxplot(x=[df[df['origin']==1]['mpg'],
               df[df['origin']==2]['mpg'],
               df[df['origin']==3]['mpg']],
            labels=['USA', 'EU', 'JAPAN'],
            vert=False)

ax1.set_title("제조국가별 연비 분포(수직 박스 플롯)")
ax2.set_title("제조국가별 연비 분포(수평 박스 플롯)")

plt.show()
```

```python
import seaborn as sn
import matplotlib.pyplot as plt

sns.set_style('darkgrid')

titanic=sns.load_dataset('titanic')

fig=plt.figure(figsize=(10, 5))

ax1=fig.add_subplot(1, 2, 1)
ax2=fig.add_subplot(1, 2, 2)

#산점도
#회기선이 있는 그래프
sns.regplot(ax=ax1,
            x='age',
            y='fare',
            data=titanic)

#회기선이 없는 그래프
sns.regplot(ax=ax2,
            x='age',
            y='fare',
            data=titanic,
            fit_reg=False)

ax1.set_title("fit_rgplot - age vs fare")
ax2.set_title("rgplot - age vs fare")

plt.show()
```
